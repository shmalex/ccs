# Charts Cheat Sheets

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = np.sin(np.array(range(1000))/100)
```

```python
fig, ax1 = plt.subplots(figsize=(14,5))
I = 300
N = 600
ax1.plot(data)
ax2 = ax1.twinx() 
ax2.hlines(0,0,N, color='black')
ax2.hlines(1,0,N, color='grey')
ax2.hlines(-1,0,N, color='grey')
ax2.hlines(-2,0,N, color='lightgrey')
ax2.hlines(2,0,N, color='lightgrey')

ax2.plot(np.cos(data)*5-5, color='orange')
# ax2.scatter(range(N),ddf.iloc[:N]['ft'], color='orange')
plt.show()
```
